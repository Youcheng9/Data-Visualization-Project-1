```python
# ~~~~~~~~ Apple Iphone Sales by Model ~~~~~~~~
import matplotlib.pyplot as plt

labels = ['iPhone 11', 'iPhone 12', 'iPhone 13',
          'iPhone 14', 'iPhone 15', 'iPhone 16', 'iPhone 17']
sales = [50.4, 51.5, 50.0, 50.9, 55.3, 55.0, 54.4]


fig = plt.figure(figsize=(10, 6))
fig.text(
    0.5, 0.000001, 
    "Question: Which model had the lowest sales?",
    ha='center',
    fontsize=17
)

plt.bar(labels, sales, color=["#bcdcf3", "#feb775", "#93eb93", "#fccaca", "#e8d0ff", "#b3e9e8", "#f0b3de"])
plt.xlabel('iPhone Models') 
plt.ylabel('Sales in Millions')
plt.title('Apple iPhone Sales by Model', fontsize=14)

plt.xticks(rotation=20)
plt.ylim(0, 70)
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.show()
```

```python
# ~~~~~~~~~ Area Chart ~~~~~~~~~

fig = plt.figure(figsize=(10, 6))
fig.text(
    0.5, 0.000001, 
    "Question: Which model had the lowest sales?",
    ha='center',
    fontsize=17
)
plt.fill_between(labels, sales, color="#5cb3f1", alpha=0.5)
plt.plot(labels, sales, marker='h', markersize=10, color="#8ab6fd")
plt.xlabel('iPhone Models')
plt.ylabel('Sales in Millions')
plt.title('Apple iPhone Sales by Model')
plt.xticks(rotation=20)
plt.ylim(0, 60)
plt.tight_layout(rect=[0, 0.1, 1, 1])


plt.show()
```
