```python
# ~~~~~~~~ Adidas Sales by Model ~~~~~~~~
import matplotlib.pyplot as plt
import numpy as np

years = [2014, 2015, 2016, 2017, 2018, 
         2019, 2020, 2021, 2022, 2023, 
         2024, 2025]
sales = [50.23, 49.75, 48.90, 50.10, 52.30,
56.00, 57.05, 54.80, 51.90, 53.10,
54.60, 53.75]


plt.figure(figsize=(10, 6))

plt.plot(years, sales, marker='o', color='orange')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Sales in Millions', fontsize=15)
plt.title('Adidas Sales (2019-2025)', fontsize=18)
plt.ylim(30, 70)
plt.xlim(2013.7, 2025.3)
plt.show()
```

```python
# ~~~~~~~~~ Area chart ~~~~~~~~~
plt.figure(figsize=(13, 4))
plt.fill_between(years, sales, color='lightcoral', alpha=0.5)
plt.plot(years, sales)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Sales in Millions', fontsize=15)
plt.title('Adidas Sales (2019-2025)', fontsize=18)
plt.ylim(30, 60)
plt.xlim(2013.7, 2025.3)
plt.show()
```
