```python
# ~~~~~~~~ Sneakers Sales by Model ~~~~~~~~
import matplotlib.pyplot as plt

labels = ['Nike Air Max', 'Adidas Ultraboost', 'Puma Suede',
          'Reebok Classic', 'Vans Old Skool', 'Converse Chuck Taylor',
          'New Balance 990', 'Onitsuka Tiger', 'Nike Air Force 1',
          'Jordan Retro', 'Nike Metcon', 'Puma Speedcat',
          'Adidas Samba']
sales = [50.4, 51.5, 50.0, 50.9, 55.3, 55.0, 54.4, 53.2, 52.8, 54.0, 53.5, 51.0, 50.7]


plt.figure(figsize=(10, 6))

plt.barh(labels, sales, color=["#bcdcf3", "#feb775", "#93eb93",
                              "#fccaca", "#e8d0ff", "#b3e9e8",
                              "#f0b3de", "#c9c9ff", "#f7d1b3",
                              "#d1f7c4", "#f0e68c", "#add8e6",
                              "#ffb6c1"])
plt.ylabel('Sneaker Models', fontsize=14) 
plt.xlabel('Sales in Millions', fontsize=14)
plt.title('Sneakers Sales by Model', fontsize=16)
plt.xticks(fontsize=12)
plt.show()
```

```python
# ~~~~~~~~~ Area Chart ~~~~~~~~~

plt.figure(figsize=(12, 8))

plt.fill_between(labels, sales, color="#60d9e7", alpha=0.5)
plt.plot(labels, sales, marker='h', markersize=10, color="#8fc7d5")
plt.xlabel('Sneaker Models', fontsize=14)
plt.ylabel('Sales in Millions', fontsize=14)
plt.title('Sneakers Sales by Model', fontsize=18)
plt.xticks(rotation=30)
plt.ylim(0, 60)
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.show()
```
