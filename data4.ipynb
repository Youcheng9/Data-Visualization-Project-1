```python
# ~~~~~~~~~~~~~ Rainfall in California ~~~~~~~~~~~~~
import matplotlib.pyplot as plt
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026]
rainfall = [15.2, 12.0, 18.5, 10.9, 20.1, 14.7, 17.6, 11.2, 19.0, 13.5, 16.8, 12.0]

plt.figure(figsize=(10, 6))

plt.fill_between(years, rainfall, color='skyblue', alpha=0.5)
plt.plot(years, rainfall, color='skyblue', marker='o')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Rainfall in Inches', fontsize=15)
plt.title('Annual Rainfall in California (2015-2026)', fontsize=18)
plt.xlim(2014.7, 2026.3)
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.show()
```

```python
# ~~~~~~ Line chart for California rainfall ~~~~~~
plt.figure(figsize=(10, 6))

plt.plot(years, rainfall, marker='o', color='skyblue')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Rainfall in Inches', fontsize=15)
plt.title('Annual Rainfall in California (2015-2026)', fontsize=18)
plt.xlim(2014.7, 2026.3)
plt.grid()
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.show()
```
