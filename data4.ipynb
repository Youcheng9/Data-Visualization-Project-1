```python
# ~~~~~~~~~~~~~ Rainfall in California ~~~~~~~~~~~~~
import matplotlib.pyplot as plt
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026]
rainfall = [15.23, 17.53, 18.55, 19.93, 18.54, 18.7, 17.6, 17.34, 19.54, 15.52, 16.86, 17.54]

plt.figure(figsize=(10, 6))
plt.bar(years, rainfall, color='skyblue')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Rainfall in Inches', fontsize=15)
plt.title('Annual Rainfall in California (2015-2026)', fontsize=18)
plt.xlim(2014.3, 2026.7)
plt.show()
```

```python
# ~~~~~~ Area chart for California rainfall ~~~~~~

plt.figure(figsize=(10, 6))

plt.fill_between(years, rainfall)
plt.plot(years, rainfall)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Rainfall in Inches', fontsize=15)
plt.title('Annual Rainfall in California (2015-2026)', fontsize=18)
plt.xlim(2014.7, 2026.3)
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.show()
```
