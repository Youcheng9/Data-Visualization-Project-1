```python
# ~~~~~~~~~ Olympic Medals 2/21/2026 Real-Time Records ~~~~~~~~~
import matplotlib.pyplot as plt

countries = ['USA', 'China', 'Japan', 
             'Norway', 'Germany', 'Australia', 'Netherlands',
             'France', 'Italy', 'Switzerland', 'Sweden', 'Austria', 'Canada', 'Great Britain']
medals = [32, 13, 24, 40, 24, 6, 20, 23, 30, 20, 16, 20, 20, 4]


plt.figure(figsize=(12, 6))

plt.bar(countries, medals, color='teal')
plt.xlabel('Countries', fontsize=15)
plt.ylabel('Number of Medals', fontsize=15)
plt.title('Olympic Medals Won by Countries in on 2/21/2026', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# ~~~~~~~ Bubble Chart ~~~~~~~~
plt.figure(figsize=(12, 6))

sizes = [s*30 for s in medals]
plt.scatter(countries, medals, s=sizes, color='green', alpha=0.5)
plt.xlabel('Countries', fontsize=15)
plt.ylabel('Number of Medals', fontsize=15)
plt.title('Olympic Medals Won by Countries in on 2/21/2026', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 45)
plt.show()
```
