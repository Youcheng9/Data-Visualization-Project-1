```python
# ~~~~~~ EV cars sales over the years ~~~~~~
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
sales = [5000, 15000, 3000, 6000, 12000, 25000, 50000, 10000, 20000, 4000, 3500]

fig = plt.figure(figsize=(10, 6))
fig.text(
    0.5, 0.000001, 
    "Question: Is the increase in EV car sales from 2015 to 2016 \ngreater than the increase from 2023 to 2024?",
    ha='center',
    fontsize=17
)

plt.plot(years, sales, marker='o', color="#C681E7")
plt.xlabel('Year', fontsize=15)
plt.ylabel('EV Car Sales', fontsize=15)
plt.xlim(2014, 2027)
plt.title('EV Car Sales Over the Years in California', fontsize=18)
plt.grid()
plt.tight_layout(rect=[0, 0.1, 1, 1])

plt.show()
```

```python
# ~~~~~~ Bubble Chart ~~~~~~
fig = plt.figure(figsize=(10, 6))
fig.text(
    0.5, 0.000001, 
    "Question: Is the increase in EV car sales from 2015 to 2016 \ngreater than the increase from 2023 to 2024?",
    ha='center',
    fontsize=17
)

sizes = [s / 10 for s in sales]
plt.scatter(years, sales, s=sizes, alpha=0.5)   
plt.xlabel('Year', fontsize=15)
plt.ylabel('EV Car Sales', fontsize=15)
plt.xlim(2014, 2027)
plt.ylim(0, 60000)
plt.title('EV Car Sales Over the Years in California', fontsize=18)
plt.grid()
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.show()
```
