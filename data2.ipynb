```python
# ~~~~~~ EV cars sales over the years ~~~~~~
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
sales = [5000, 15000, 3000, 6000, 12000, 25000, 50000, 10000, 20000, 4000, 3500]
plt.plot(years, sales, marker='o')
plt.xlabel('Year')
plt.ylabel('EV Car Sales')
plt.xlim(2014, 2027)
plt.title('EV Car Sales Over the Years in California')
plt.grid()
plt.show()
```

```python
# ~~~~~~ Bubble Chart ~~~~~~
sizes = [s / 10 for s in sales]
plt.scatter(years, sales, s=sizes, alpha=0.5)   
plt.xlabel('Year')
plt.ylabel('EV Car Sales')
plt.xlim(2014, 2027)
plt.ylim(0, 60000)
plt.title('EV Car Sales Over the Years in California')
plt.grid()
plt.show()
```
