```python
# ~~~~~~ EV cars sales over the years ~~~~~~
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
sales = [12345, 13423, 12325, 16542, 18165, 21034, 20133, 19342, 18001, 20312, 20421]

plt.figure(figsize=(15, 6))
plt.plot(years, sales, color='red')
plt.xlabel('Year', fontsize=15)
plt.ylabel('EV Car Sales', fontsize=15)
plt.ylim(10000, 25000)
plt.xlim(2014.5, 2025.5)
plt.title('EV Car Sales Over the Years in California', fontsize=18)
plt.show()
```

```python
# ~~~~~~ Bubble Chart ~~~~~~
plt.figure(figsize=(10, 6))


sizes = [s / 10 for s in sales]
plt.scatter(years, sales, s=sizes, alpha=0.5)   
plt.xlabel('Year', fontsize=15)
plt.ylabel('EV Car Sales', fontsize=15)
plt.xlim(2014.5, 2025.5)
plt.ylim(10000, 25000)
plt.title('EV Car Sales Over the Years in California', fontsize=18)
plt.show()
```
