```python
import matplotlib.pyplot as plt
import numpy as np


netlifx = 15.95
hulu = 14.99
amazon_prime = 15.45
disney_plus = 15.70
hbo_max = 14.99

labels = ['Netflix', 'Hulu', 'Amazon Prime', 'Disney Plus', 'HBO Max']
values = [netlifx, hulu, amazon_prime, disney_plus, hbo_max]

plt.bar(labels, values, color=["#e83e3e", "#ffdd95", "#5bacbb", "#c986e6", "#7dca6e"])
plt.xlabel('Streaming Services')
plt.ylabel('Subscription Cost ($)')
plt.title('Subscription Costs of Streaming Services')
plt.xticks(rotation=10)
plt.ylim(0, 30)
plt.show()
```

```python
# ~~~~~~~~~~ Marker Chart ~~~~~~~~~~
plt.plot(labels, values, 
         marker='o',         
         markersize=15,       
         markerfacecolor="#f66959", 
         markeredgecolor='#e83e3e', 
         linestyle=''      
        )
plt.xlabel('Streaming Services')
plt.ylabel('Subscription Cost ($)')
plt.title('Subscription Costs of Streaming Services')
plt.ylim(0, 30)
plt.grid(True)
plt.show()
```
