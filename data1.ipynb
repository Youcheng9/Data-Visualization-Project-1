```python
# ~~~~~~~ Degree Popularity at HK State Polytechnic ~~~~~~~

import matplotlib.pyplot as plt
import numpy as np

labels = ['Business', 'Engineering', 'Architecture', 'Computer Science',
          'Education', 'Biology', 'Psychology', 'Nursing', 'Economics',
          'Political Science', 'Sociology', 'History', 'Philosophy',
          'Art History', 'Anthropology']

values = [4014, 4945, 4403, 5265, 4541, 4034, 4853, 5120, 4756, 5644, 5586, 5453, 5364, 5455, 4352]

plt.figure(figsize=(10, 6))


plt.bar(labels, values, color=["#e83e3e", "#ffdd95", "#5bacbb", "#c986e6", "#7dca6e"])
plt.xlabel('Degrees', fontsize=15)
plt.ylabel('Number of Students', fontsize=15)
plt.title('Degree Popularity at HK State Polytechnic', fontsize=20)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 6000)
plt.show()
```

```python
# ~~~~~~~~~~ Bubble Chart ~~~~~~~~~
plt.figure(figsize=(17, 8))

plt.scatter(labels, values, s=np.array(values) * 0.5, alpha=0.5, color="#d88cf0")
plt.xlabel('Degrees', fontsize=19)
plt.ylabel('Number of Students', fontsize=19)
plt.title('Degree Popularity at HK State Polytechnic', fontsize=28)
plt.xlim(-1, len(labels))
plt.xticks(rotation=30)
plt.ylim(3000, 6000)
plt.show()
```
