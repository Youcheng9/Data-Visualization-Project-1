```python
import matplotlib.pyplot as plt
import numpy as np


netlifx = 15.95
hulu = 14.99
amazon_prime = 15.45
disney_plus = 15.70
hbo_max = 14.99

labels = ['Netflix', 'Hulu', 'Amazon Prime', 'Disney Plus', 'HBO Max']
values = [netlifx, hulu, amazon_prime, disney_plus, hbo_max]


plt.figure(figsize=(10, 6))


plt.bar(labels, values, color=["#e83e3e", "#ffdd95", "#5bacbb", "#c986e6", "#7dca6e"])
plt.xlabel('Streaming Services', fontsize=15)
plt.ylabel('Subscription Cost ($)', fontsize=15)
plt.title('Subscription Costs of Streaming Services', fontsize=20)
plt.xticks()
plt.ylim(0, 30)
plt.show()
```

```python
# ~~~~~~~~~~ Marker Chart ~~~~~~~~~~
plt.figure(figsize=(10, 6))

plt.plot(labels, values, 
         marker='o',         
         markersize=15,       
         markerfacecolor="#f66959", 
         markeredgecolor='#e83e3e'     
        )
plt.xlabel('Streaming Services', fontsize=15)
plt.ylabel('Subscription Cost ($)', fontsize=15)
plt.title('Subscription Costs of Streaming Services', fontsize=18)
plt.ylim(0, 30)
plt.grid(True)
plt.show()
```
