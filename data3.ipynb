```python
# ~~~~~~~~ Cancer Death Rates in 30-60s Age Group Over the Years in the US~~~~~~~~
import matplotlib.pyplot as plt
import numpy as np

labels = ['Lung', 'Breast', 'Prostate', 'Colorectal', 'Pancreatic', 'Skin']
values = {
    '30s': [5000, 18000, 200, 12000, 2000, 1500],
    '40s': [10000, 15000, 500, 8000, 4000, 3000],
    '50s': [14000, 20000, 1000, 4000, 8000, 5000],
    '60s': [20000, 9000, 5000, 3700, 12000, 8000]
}
x = np.arange(len(labels))
width = 0.2

fig, ax = plt.subplots(figsize=(12, 7))
fig.text(
    0.5, 0.000001, 
    "Question: Which age group has the highest cancer death rates?",
    ha='center',
    fontsize=17
)

for i, (age_group, death_rates) in enumerate(values.items()):
    ax.bar(x + i*width, death_rates, width, label=age_group)

ax.set_xticks(x + width*1.5)
ax.set_xticklabels(labels, rotation=30)
ax.set_xlabel('Cancer Types', fontsize=15)
ax.set_ylabel('Death Rates', fontsize=15)
ax.set_title('Cancer Death Rates by Age Group', fontsize=18)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.show()
```

```python
# ~~~~~~ Heatmap ~~~~~~
data = np.array(list(values.values()))

fig = plt.figure(figsize=(10, 5))
fig.text(
    0.5, 0.000001, 
    "Question: Which age group has the highest cancer death rates?",
    ha='center',
    fontsize=17
)

plt.imshow(data, cmap='BuPu', aspect='auto')
plt.colorbar(label='Death Rates')
plt.xticks(ticks=np.arange(len(labels)), labels=labels, rotation=30)
plt.yticks(ticks=np.arange(len(values)), labels=list(values.keys()))
plt.xlabel('Cancer Types', fontsize=15)
plt.ylabel('Age Groups', fontsize=15)
plt.title('Heatmap of Cancer Death Rates by Age Group', fontsize=18)
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.show()
```
